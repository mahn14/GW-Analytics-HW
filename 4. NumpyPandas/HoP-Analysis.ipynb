To look at whether the norms are significantly different, we can use a t-test looking at each individual's total purchasing values and comparing through a t-test.  
H0: The two age groups have identical averages  
H1: The two age groups have different averages 

```python
def diff_test(i, j):
    df1 = np.sum(data[data['Age Cat'] == j].groupby('SN')['Price'])
    df2 = np.sum(data[data['Age Cat'] == i].groupby('SN')['Price'])
    return(stat.ttest_ind(df1, df2))
```

```python
[diff_test(i, 6)[1] > 0.05 for i in range(6)]
```

Running the test across all relevant age groups, omitting 40+ since the sample size is too small, we reject the null hypothesis for only the 15-24 age groups.  
However, it is important to note that tests comparing the <10 and 15-29 groups, for example, fail to reject the null hypothesis.
